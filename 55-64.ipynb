```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import linregress
```

```python
all_data = pd.read_csv("project_1_resources/all_data_v3.csv")
all_data
```

```python
fifty_five_to_sixty_four_df = all_data.loc[(all_data['Age Group_x'] == '55 to 64 years') & (all_data['Age Group_y'] == '55 to 64 years') ]
fifty_five_to_sixty_four_df
```

```python
fifty_five_to_sixty_four_df_male = fifty_five_to_sixty_four_df.loc[(fifty_five_to_sixty_four_df['Gender_x' ] == 'Male')]
fifty_five_to_sixty_four_df_male = fifty_five_to_sixty_four_df_male[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
fifty_five_to_sixty_four_df_male.columns =['Unique ID','Year','Male Avg Sleep','Male Avg income']
fifty_five_to_sixty_four_df_male

fifty_five_to_sixty_four_df_female = fifty_five_to_sixty_four_df.loc[(fifty_five_to_sixty_four_df['Gender_x' ] == 'Female')]
fifty_five_to_sixty_four_df_female = fifty_five_to_sixty_four_df_female[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
fifty_five_to_sixty_four_df_female.columns =['Unique ID','Year','Female Avg Sleep','Female Avg income']
fifty_five_to_sixty_four_df_female

fifty_five_to_sixty_four_df_mvf = pd.merge(fifty_five_to_sixty_four_df_male,fifty_five_to_sixty_four_df_female, on = 'Year', how = 'inner')
fifty_five_to_sixty_four_df_mvf
```

```python

fifty_five_to_sixty_four_df_mvf['Year'] = fifty_five_to_sixty_four_df_mvf['Year'].astype(str)
ax1 = fifty_five_to_sixty_four_df_mvf.plot.bar('Year',['Male Avg income','Female Avg income'])

# make secondary y axis
ax2 = ax1.twinx()
#secondary y axis label

#make line charts
fifty_five_to_sixty_four_df_mvf.plot('Year', 'Male Avg Sleep',ax=ax2,color = 'blue')
fifty_five_to_sixty_four_df_mvf.plot('Year', 'Female Avg Sleep',ax=ax2,color = 'orange')


ax1.set_ylabel('Average Annual Income( In Dollars)')
ax2.set_ylabel('Average Hours of Sleep')
plt.title("Gender Comparison of Sleep Habits and Annual Income for Ages 55 - 64")

ax1.legend(["Male Average Annual Income", "Female Average Annual Income"]);
ax2.legend(["Male Average Hours of Sleep", "Female Average Hours of Sleep"], loc = "lower right")
plt.show()
```

```python
fifty_five_to_sixty_four_avg_sleep = fifty_five_to_sixty_four_df['Avg hrs per day sleeping']
fifty_five_to_sixty_four_avg_income = fifty_five_to_sixty_four_df['Mean income']

(slope, intercept, rvalue, pvalue, stderr) = linregress(fifty_five_to_sixty_four_avg_sleep, fifty_five_to_sixty_four_avg_income )
regress_values =  fifty_five_to_sixty_four_avg_sleep * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(fifty_five_to_sixty_four_avg_sleep,fifty_five_to_sixty_four_avg_income)
plt.plot(fifty_five_to_sixty_four_avg_sleep,regress_values,"r-")


plt.xlabel('Average Hours of Sleep')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income vs Average Hours of Sleep Regression Plot for Ages 55 -64')
plt.annotate(line_eq,(9,10),fontsize=15,color="red")

plt.show()
print(f'The r-value is:{rvalue**2}')
```
