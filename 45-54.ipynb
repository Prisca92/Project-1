```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import linregress
```

```python
all_data = pd.read_csv("project_1_resources/all_data_v3.csv")
all_data
```

```python
forty_five_to_fifty_four_df = all_data.loc[(all_data['Age Group_x'] == '45 to 54 years') & (all_data['Age Group_y'] == '45 to 54 years') ]
forty_five_to_fifty_four_df
```

```python
forty_five_to_fifty_four_df
forty_five_to_fifty_four_df_male = forty_five_to_fifty_four_df.loc[(forty_five_to_fifty_four_df['Gender_x' ] == 'Male')]
forty_five_to_fifty_four_df_male = forty_five_to_fifty_four_df_male[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
forty_five_to_fifty_four_df_male.columns =['Unique ID','Year','Male Avg Sleep','Male Avg income']
forty_five_to_fifty_four_df_male


forty_five_to_fifty_four_df_female = forty_five_to_fifty_four_df.loc[(forty_five_to_fifty_four_df['Gender_x' ] == 'Female')]
forty_five_to_fifty_four_df_female = forty_five_to_fifty_four_df_female[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
forty_five_to_fifty_four_df_female.columns =['Unique ID','Year','Female Avg Sleep','Female Avg income']
forty_five_to_fifty_four_df_female

forty_five_to_fifty_four_df_mvf = pd.merge(forty_five_to_fifty_four_df_male,forty_five_to_fifty_four_df_female, on = 'Year', how = 'inner')
forty_five_to_fifty_four_df_mvf
```

```python

forty_five_to_fifty_four_df_mvf['Year'] = forty_five_to_fifty_four_df_mvf['Year'].astype(str)
ax1 = forty_five_to_fifty_four_df_mvf.plot.bar('Year',['Male Avg income','Female Avg income'])

# make secondary y axis
ax2 = ax1.twinx()
#secondary y axis label

#make line charts
forty_five_to_fifty_four_df_mvf.plot('Year', 'Male Avg Sleep',ax=ax2,color = 'blue')
forty_five_to_fifty_four_df_mvf.plot('Year', 'Female Avg Sleep',ax=ax2,color = 'orange')


ax1.set_ylabel('Average Annual Income( In Dollars)')
ax2.set_ylabel('Average Hours of Sleep')
plt.title("Gender Comparison of Sleep Habits and Annual Income for Ages 45 - 54")

ax1.legend(["Male Average Annual Income", "Female Average Annual Income"]);
ax2.legend(["Male Average Hours of Sleep", "Female Average Hours of Sleep"], loc = "lower right")
plt.show()
```

```python
forty_five_to_fifty_four_avg_sleep = forty_five_to_fifty_four_df['Avg hrs per day sleeping']
forty_five_to_fifty_four_avg_income = forty_five_to_fifty_four_df['Mean income']

(slope, intercept, rvalue, pvalue, stderr) = linregress(forty_five_to_fifty_four_avg_sleep ,forty_five_to_fifty_four_avg_income)
regress_values =  forty_five_to_fifty_four_avg_sleep * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(forty_five_to_fifty_four_avg_sleep,forty_five_to_fifty_four_avg_income)
plt.plot(forty_five_to_fifty_four_avg_sleep ,regress_values,"r-")


plt.xlabel('Average Hours of Sleep')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income vs Average Hours of Sleep Regression Plot for Ages 45 -54')
plt.annotate(line_eq,(9,10),fontsize=15,color="red")

plt.show()
print(f'The r-value is:{rvalue**2}')
```
