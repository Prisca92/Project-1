```python
#dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import linregress
```

```python
all_data = pd.read_csv("project_1_resources/all_data_v3.csv")
all_data
```

```python
fifteen_to_twenty_four_df = all_data.loc[(all_data['Age Group_x'] == '15 to 24 years') & (all_data['Age Group_y'] == '15 to 24 years') ]
fifteen_to_twenty_four_df
```

```python
fifteen_to_twenty_four_df
fifteen_to_twenty_four_df_male = fifteen_to_twenty_four_df.loc[(fifteen_to_twenty_four_df['Gender_x' ] == 'Male')]
fifteen_to_twenty_four_df_male = fifteen_to_twenty_four_df_male[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
fifteen_to_twenty_four_df_male.columns =['Unique ID','Year','Male Avg Sleep','Male Avg income']
fifteen_to_twenty_four_df_male


fifteen_to_twenty_four_df_female = fifteen_to_twenty_four_df.loc[(fifteen_to_twenty_four_df['Gender_x' ] == 'Female')]
fifteen_to_twenty_four_df_female = fifteen_to_twenty_four_df_female[['Unique ID','Year_x','Avg hrs per day sleeping','Mean income']]
fifteen_to_twenty_four_df_female.columns =['Unique ID','Year','Female Avg Sleep','Female Avg income']
fifteen_to_twenty_four_df_female

fifteen_to_twenty_four_df_mvf = pd.merge(fifteen_to_twenty_four_df_male,fifteen_to_twenty_four_df_female, on = 'Year', how = 'inner')
fifteen_to_twenty_four_df_mvf
```

```python
fifteen_to_twenty_four_df_mvf['Year'] = fifteen_to_twenty_four_df_mvf['Year'].astype(str)
ax1 = fifteen_to_twenty_four_df_mvf.plot.bar('Year',['Male Avg income','Female Avg income'])

# make secondary y axis
ax2 = ax1.twinx()
#secondary y axis label

#make line charts
fifteen_to_twenty_four_df_mvf.plot('Year', 'Male Avg Sleep',ax=ax2,color = 'blue')
fifteen_to_twenty_four_df_mvf.plot('Year', 'Female Avg Sleep',ax=ax2,color = 'orange')


ax1.set_ylabel('Average Annual Income( In Dollars)')
ax2.set_ylabel('Average Hours of Sleep')
plt.title("Gender Comparison of Sleep Habits and Annual Income for Ages 15 - 24")

ax1.legend(["Male Average Annual Income", "Female Average Annual Income"]);
ax2.legend(["Male Average Hours of Sleep", "Female Average Hours of Sleep"], loc = "lower right")
plt.show()
```

```python
fifteen_to_twenty_four_avg_sleep = fifteen_to_twenty_four_df['Avg hrs per day sleeping']
fifteen_to_twenty_four_avg_income = fifteen_to_twenty_four_df['Mean income']

(slope, intercept, rvalue, pvalue, stderr) = linregress(fifteen_to_twenty_four_avg_sleep,fifteen_to_twenty_four_avg_income)
regress_values =  fifteen_to_twenty_four_avg_sleep * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(fifteen_to_twenty_four_avg_sleep,fifteen_to_twenty_four_avg_income)
plt.plot(fifteen_to_twenty_four_avg_sleep ,regress_values,"r-")


plt.xlabel('Average Hours of Sleep')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income vs Average Hours of Sleep Regression Plot for Ages 15 -24')
plt.annotate(line_eq,(9,10),fontsize=15,color="red")

plt.show()
print(f'The r-value is:{rvalue**2}')
```

```python
fifteen_to_twenty_four_avg_sleep = fifteen_to_twenty_four_df['Avg hrs per day sleeping']
fifteen_to_twenty_four_avg_income_twenty_two = fifteen_to_twenty_four_df['2022\ndollars']

(slope, intercept, rvalue, pvalue, stderr) = linregress(fifteen_to_twenty_four_avg_sleep,fifteen_to_twenty_four_avg_income)
regress_values =  fifteen_to_twenty_four_avg_sleep * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(fifteen_to_twenty_four_avg_sleep,fifteen_to_twenty_four_avg_income_twenty_two)
plt.plot(fifteen_to_twenty_four_avg_sleep ,regress_values,"r-")


plt.xlabel('Average Hours of Sleep')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income vs Average Hours of Sleep Regression Plot for Ages 15 -24')
plt.annotate(line_eq,(9,10),fontsize=15,color="red")

plt.show()
print(f'The r-value is:{rvalue**2}')
```

```python
fifteen_to_twenty_four_df
fifteen_to_twenty_four_df_male_twenty_two = fifteen_to_twenty_four_df.loc[(fifteen_to_twenty_four_df['Gender_x' ] == 'Male')]
fifteen_to_twenty_four_df_male_twenty_two = fifteen_to_twenty_four_df_male_twenty_two[['Unique ID','Year_x','Avg hrs per day sleeping','2022\ndollars']]
fifteen_to_twenty_four_df_male_twenty_two.columns =['Unique ID','Year','Male Avg Sleep','Male Avg income(in 2022 Dollars)']
fifteen_to_twenty_four_df_male


fifteen_to_twenty_four_df_female_twenty_two = fifteen_to_twenty_four_df.loc[(fifteen_to_twenty_four_df['Gender_x' ] == 'Female')]
fifteen_to_twenty_four_df_female_twenty_two = fifteen_to_twenty_four_df_female_twenty_two[['Unique ID','Year_x','Avg hrs per day sleeping','2022\ndollars']]
fifteen_to_twenty_four_df_female_twenty_two.columns =['Unique ID','Year','Female Avg Sleep','Female Avg income(in 2022 Dollars)']
fifteen_to_twenty_four_df_female_twenty_two

fifteen_to_twenty_four_df_mvf_twenty_two = pd.merge(fifteen_to_twenty_four_df_male_twenty_two,fifteen_to_twenty_four_df_female_twenty_two, on = 'Year', how = 'inner')
fifteen_to_twenty_four_df_mvf_twenty_two
```

```python
fifteen_to_twenty_four_df_mvf_twenty_two['Year'] = fifteen_to_twenty_four_df_mvf['Year'].astype(str)
ax1 = fifteen_to_twenty_four_df_mvf_twenty_two.plot.bar('Year',['Male Avg income(in 2022 Dollars)','Female Avg income(in 2022 Dollars)'])

# make secondary y axis
ax2 = ax1.twinx()
#secondary y axis label

#make line charts
fifteen_to_twenty_four_df_mvf_twenty_two.plot('Year', 'Male Avg Sleep',ax=ax2,color = 'blue')
fifteen_to_twenty_four_df_mvf_twenty_two.plot('Year', 'Female Avg Sleep',ax=ax2,color = 'orange')


ax1.set_ylabel('Average Annual Income( In 2022 Dollars)')
ax2.set_ylabel('Average Hours of Sleep')
plt.title("Gender Comparison of Sleep Habits and Annual Income for Ages 15 - 24")

ax1.legend(["Male Average Annual Income", "Female Average Annual Income"]);
ax2.legend(["Male Average Hours of Sleep", "Female Average Hours of Sleep"], loc = "lower right")
plt.show()
```
